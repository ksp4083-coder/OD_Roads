## 모델 학습

- YOLOv8 세부 모델 5개를 학습하고 평가한 뒤, mAP (IoU threshold = 0.5) 기준으로 가장 높은 정확도를 보이는 모델을 7일 동안 재학습

- 모델을 오래 학습시킬 수록 정확도는 높아지지만, 학습 데이터셋에서는 높은 성능을 보이고 검증 데이터셋에서는 낮은 성능을 보이는 **모델 과적합**(Overfitting)을 우려함

- 대용량 데이터셋(COCO dataset)에서 **사전 학습(pre-trained)된 모델의 높은 일반화 성능**을 활용하여 오랫동안 모델을 학습시켜도 과적합이 발생하지 않을 것을 가정

- 과적합은 학습 결과로 제공되는 **학습/검증 손실 그래프**(train/loss graph)가 안정적으로 수렴하는지 확인하고 판단함

- 학습 코드 출처

    - https://docs.ultralytics.com/ko/modes/train/#resuming-interrupted-trainings 

    - https://velog.io/@spacenet/YOLO-v8-easy-to-use 

- 참고

    - 아래 코드는 RTX 5090 학습 성능 평가를 위한 코드이며 공모전 코드 예는 "2024_summer_train.ipynb"와 같음

    - 이외의 학습 코드는 날아간 관계로 양해 바람
    
    - 공모전 진행 당시 모델은 DSC 공유대학 GPU (A100) 서버에서 학습시킴
